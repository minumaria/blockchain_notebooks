# Inter Planetory File System (IPFS)

### This notebook will cover the following:

* Installing IPFS on local machine
* Initializing an IPFS node
* Starting the IPFS daemon
* Adding files to IPFS
* Pinning files to local machine
* Name publishing using IPNS

Download ipfs-go from https://dist.ipfs.io and extract the tar ball

> wget https://dist.ipfs.io/go-ipfs/v0.4.10/go-ipfs_v0.4.10_linux-amd64.tar.gz <br>
> tar -xvzf ./go-ipfs* <br>

Install ipfs-go on local machine by running the install.sh in executable mode

> cd go-ipfs && sudo . /install.sh <br>

Initialize the ipfs on local machine and make a note of the node hash generated

> ipfs init <br>

Simply start the IPFS daemon

> Ipfs daemon <br>

Now open a new terminal and create a single page website

> nano test.html

Add the single page website html file to IPFS network

> ipfs add -r test.html <br>

To keep the file always accessible let's pin to our local machine

> ipfs pin add -r file_hash <br>

Now let's publish our website

> ipfs name publish file_hash

### Let's look for our single page static website now

visit localhost:8080/ipfs/file_hash

visit localhost:8080/ipns/peer_id

visit gateway.ipfs.io/ipfs/file_hash

visit gateway.ipfs.io/ipns/file_hash

### Now stop the local daemon

### Also delete the static website file

### visit the gateway.ipfs.io links again and check

### Get a list of all connected peers

> ipfs swarm peers

### Search for a particular peer

> ipfs dht findpeer peer_hash

### IPFS Bootstrap

A list of peers with which the IPFS daemon learns about other peers on the network.

> ipfs bootstrap list
